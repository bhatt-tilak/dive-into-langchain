### Documentation for future reference 

- initialize project using uv
uv init

- create requirements.txt file
add following 

```
langchain
langchain_community
langchain-openai
langchain-groq
python-dotenv
langchain-google-genai
langchain-ollama
```

- initialize virtual environment for python

```
uv venv
```

- start virtual environment
check your directory and os

```
source ./.venv/bin/active 
```

- add ipykernel

```
uv add ipykernel
```

- install from requirements.txt using uv

````
uv add -r requirements.txt
````


- loading environment variables 
- add a .env file 
- add all files in format below
- ENV_KEY_NAME="<Key-Here>"

- load key like this 

```
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```
