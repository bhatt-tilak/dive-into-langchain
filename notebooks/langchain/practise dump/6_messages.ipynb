```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    "llama3.2:3b",
    model_provider="ollama"
)

model


# messages are data structures for llms
# fundamental unit of context
# they represent state of conversation
# message has , role, content, metadata

# Human Message , AI message, System Message, Tool Message

# text prompts 



# this is just text prompt
# model.invoke("What is ai?").content

# message prompt 
message_prompt = {
    "role": "user",
    "content": "What is AI?"
}

# System Message - How the model should behave
# Human Message - What the user is asking
# AI Message - What the model is responding with
# Tool Message - What the model is asking the tool to do


from langchain.messages import HumanMessage, AIMessage, SystemMessage

messages = [
    SystemMessage("You are a poet."),
    HumanMessage("Write a 20 words poem about the moon."),
]
messages
response = model.invoke(messages)
response
```

```python
system_message = SystemMessage("You are a helpful coding assistant.")

messages = [
    system_message,
    HumanMessage("Write a python function to add two numbers."),
]

response = model.invoke(messages)
response.content
```

```python
# detailed system message 

system_message = SystemMessage(
    """
        You are a senior python developer with 10 years of experience. 
        You are an expert in writing clean and efficient code. 
        You have a deep understanding of python libraries and frameworks. 
        You are also a great teacher and can explain complex concepts in a simple way.
    """
)

messages = [
    system_message,
    HumanMessage("How to create rest api."),
]

reponse = model.invoke(messages)
```
